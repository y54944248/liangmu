> - isalnum() 中是不能包含空格，否则会返回False

```python
def check(a):
    """
    check Password
    Parameters:
    --------
        a:Need check password.
    Returns:
    -------
        None.
    Note:
    ----
        1.Must use for loop.
        2.必须写注释
    """
    N1,N2,N3 = 0,0,0
    for i in a:
        if i.isupper():
            N1+=1
        elif i.islower():
            N2+=1
        elif i.isdigit():
            N3+=1
    if N1 == 0:
        print("密码必须有大写字母")
    if N2 == 0:
        print("密码必须有小写字母")
    if N3 == 0:
        print("密码必须有数字")
```

```python
a = "12346ff"
check(a)
```

```python
def countLetters(s):
     """
    定义一个函数，通过调用函数实现统计字符串中字母出现的次数
    Parameters:
    --------
        a:字母出现的形式只有大写或者小写两种形式，所以判断字母是否是大小写，如果是大写，自加一；
        如果小写，自加一。
    Note:
    ----
        1.使用isupper，islower判断字母.
    """
    N=0
    for i in s:
        if i.isupper(): #检验大写
            N+=1
        if i.islower(): #检验小写
            N+=1
    print(N)
```

```python
s="1234sd56"
countLetters(s)
```
